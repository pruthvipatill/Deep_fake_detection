```python
import cv2
from mtcnn import MTCNN
import matplotlib.pyplot as plt
import numpy as np
```

```python
detector = MTCNN()

video_path = r'C:\Users\PRUTHVI\OneDrive\Desktop\RBL\Deep_Fake_Detection\face_detection.mp4'
cap = cv2.VideoCapture(video_path)
```

```python
success , frame = cap.read()
if success:

    frame_rgb = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)

    faces = detector.detect_faces(frame_rgb)

    for face in faces:
        x,y,width,height = face['box']
        cv2.rectangle(frame_rgb,(x,y),(x+width,y+height),(0,255,0),5)
    
    plt.imshow(frame_rgb)
    plt.title('Face Extraction via MTCNN')
    plt.axis('off')
    plt.show()
else:
    print('Could not read the video bruh')
```

```python
face_crop = frame_rgb[y:y+height,x:x+width]

gray_face = cv2.cvtColor(face_crop,cv2.COLOR_RGB2GRAY)

f = np.fft.fft2(gray_face)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift)+1)
```

```python
plt.figure(figsize=(8,4))

plt.subplot(121)
plt.imshow(face_crop)
plt.title('Extracted Face(spatial Domain)')
plt.axis('off')

plt.subplot(122)
plt.imshow(magnitude_spectrum,cmap='magma')
plt.title('Frequency spectrum (Frequency stream)')
plt.axis('off')

plt.suptitle('Pipeline step 2: Modality Extraction', fontsize = 14 , fontweight = 'bold')
plt.show()
```
